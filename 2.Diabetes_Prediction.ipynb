```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_ds=pd.read_csv("/content/drive/MyDrive/ML Workshop/dataset/diabetes.csv")
```

```python
diabetes_ds.head()
```

```python
diabetes_ds.shape
```

```python
diabetes_ds.describe()
```

```python
diabetes_ds["Outcome"].value_counts()
```

0 is Non-Diabetic        
1 is Diabetic

```python
diabetes_ds.groupby("Outcome").mean()
```

```python
#separating dependent and independent columns

X=diabetes_ds.drop(columns="Outcome",axis=1)
Y=diabetes_ds["Outcome"]

print(X)
print(Y)
```

```python
print(diabetes_ds.std())  #gives us the deviation of data
```

```python
print(X.std())
```

```python
#We need to make the data have as less deviation as possible, for accurate model

scaler=StandardScaler()
scaler.fit(X)   #We only need to Standardize the deviation of input(Independent variables)

std_data=scaler.transform(X)
print(std_data)
```

```python
print(std_data.std())      #This is the standardized data of input
print(X.std())             #Previous un-standardized data
```

```python
X=std_data                   #New Independent variable
Y=diabetes_ds["Outcome"]

print(X.std())
```

Train-Test Split

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(Y_train.shape, Y_test.shape)
```

Training the model

```python
classifier=svm.SVC(kernel="linear")
```

```python
#training the Support Vector Machine classifier
classifier.fit(X_train, Y_train)
```

Evaluation

```python
#accuracy of training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)

print("Accuracy of training data:",training_data_accuracy)
```

```python
#accuracy of testing data
X_test_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction, Y_test)

print("Accuracy of testing data:",testing_data_accuracy)
```

Checking the model

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

arr=np.array(input_data)
arr_reshaped=arr.reshape(1,-1)

#standardize the input data
std_data=scaler.transform(arr_reshaped)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
  print("Not Diabetic")
else:
  print("Diabetic")
```
