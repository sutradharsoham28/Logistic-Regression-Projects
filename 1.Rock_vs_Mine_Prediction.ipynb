**Importing Dependancies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression   #used for binary prediction(rock or sea mine for submarines)
from sklearn.metrics import accuracy_score
```

```python
 #loading dataset

sonar_data=pd.read_csv("/content/drive/MyDrive/ML Workshop/dataset/sonar data.csv",header=None)
```

```python
#separating the independent(feature_names) and dependent(target_array) variables
X=sonar_data.drop(columns=60,axis=1)  #This will store the first 60 columns(the independent variables)
Y=sonar_data[60]   #This will store the last column(the dependent variable)
```

```python
print(X)
print(Y)
```

Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=1)
#test_size=0.1  means that 10% of the data will go for training
#random_state=1 means the shuffling of rows(different shuffling means different accuracy)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

Model Training-->Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
print(X_train_prediction)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data:",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)

print("Accuracy on test data:",test_data_accuracy)
```
